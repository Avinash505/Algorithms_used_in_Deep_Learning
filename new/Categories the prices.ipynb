```python
def categorize_prices(prices):
    # Step 1: Sort the data in ascending order
    sorted_prices = np.sort(prices)

    # Step 2: Calculate the quartiles (percentiles)
    q1 = np.percentile(sorted_prices, 25)
    q2 = np.percentile(sorted_prices, 50)
    q3 = np.percentile(sorted_prices, 75)

    # Step 3: Define the categories
    categories = {
        'Category 1 (min-Q1)': f"{sorted_prices[0]}-{q1}",
        'Category 2 (Q1-Q2)': f"{q1 + 1}-{q2}",
        'Category 3 (Q2-Q3)': f"{q2 + 1}-{q3}",
        'Category 4 (Q3-max)': f"{q3 + 1}-{sorted_prices[-1]}"
    }
    return categories

result = categorize_prices(features)
for category, price_range in result.items():
    print(f"{category}: {price_range}")
data['results'] = pd.qcut(features, q=4, labels=['Category 1 (min-Q1)', 'Category 2 (Q1-Q2)', 'Category 3 (Q2-Q3)', 'Category 4 (Q3-max)'])
```

```python
import numpy as np

def categorize_prices(prices, user_input):
    # Step 1: Sort the data in ascending order
    sorted_prices = np.sort(prices)

    # Step 2: Calculate the quartiles (percentiles)
    q1 = np.percentile(sorted_prices, 35)
    q2 = np.percentile(sorted_prices, 60)
    q3 = np.percentile(sorted_prices, 85)
    

    # Step 3: Define the categories
    categories = {
        'Category 1 (min-Q1)': f"{sorted_prices[0]}-{q1}",
        'Category 2 (Q1-Q2)': f"{q1 + 1}-{q2}",
        'Category 3 (Q2-Q3)': f"{q2 + 1}-{q3}",
        'Category 4 (Q3-max)': f"{q3 + 1}-{sorted_prices[-1]}"
    }

    # Step 4: Find the category to which the user input belongs
    input_category = None
    user_price = float(user_input)
    
    if user_price <= q1:
        input_category = 'Category 1 (min-Q1)'
    elif q1 < user_price <= q2:
        input_category = 'Category 2 (Q1-Q2)'
    elif q2 < user_price <= q3:
        input_category = 'Category 3 (Q2-Q3)'
    else:
        input_category = 'Category 4 (Q3-max)'

    return categories, input_category

user_input_price = input("Enter the price: ")
result, user_category = categorize_prices(features, user_input_price)

print("Categories:")
for category, price_range in result.items():
    print(f"{category}: {price_range}")

print(f"User input belongs to: {user_category}")
```

```python
data = data.sort_values(by='price')
Q1 = data['price'].quantile(0.25)
Q2 = data['price'].quantile(0.50)
Q3 = data['price'].quantile(0.75)

def categorize_price(price):
    if price <= Q1:
        return 'min-Q1'
    elif price <= Q2:
        return 'Q1-Q2'
    elif price <= Q3:
        return 'Q2-Q3'
    else:
        return 'Q3-max'

data['percentage_price'] = data['price'].apply(categorize_price)

# Count each value and add to the percentage_price categories
counts = data['percentage_price'].value_counts()
data['counts_in_category'] = data['percentage_price'].map(counts)
```
