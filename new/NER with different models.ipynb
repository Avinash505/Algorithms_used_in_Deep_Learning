# Stanford NER:

Stanford NER is a widely used NER system that implements various NER models, such as 3-class (person, organization, location), 4-class (person, organization, location, miscellaneous), and 7-class (person, organization, location, date, time, percent, miscellaneous).

```python
from nltk.tag import StanfordNERTagger

jar = "stanford-ner-2015-04-20/stanford-ner-3.5.2.jar"
model = "stanford-ner-2015-04-20/classifiers/english.all.3class.distsim.crf.ser.gz"

st = StanfordNERTagger(model, jar, encoding='utf8')
```

```python
# Create an empty list to store the entities
entities = []
# Iterate over the texts in the DataFrame
for text in df['text']:
    # Split the text into tokens
    tokens = text.split()
    # Perform NER on the tokens
    tagged_entities = st.tag(tokens)
    # Append the tagged entities to the list
    entities.append(tagged_entities)
```

```python
# Iterate over the entities list and print the results
for tagged_entities in entities:
    for entity in tagged_entities:
        print(entity)
```

```python
df_entities = pd.DataFrame(entities)

# Save the DataFrame to a CSV file
df_entities.to_csv('main_entities.csv', index=False)
```
