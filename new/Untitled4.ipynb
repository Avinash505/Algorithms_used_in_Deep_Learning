```python
def convert_to_float(sqft):
    try:
        return float(sqft)
    except ValueError:
        try:
            start, end = map(float, sqft.split('-'))
            return (start + end) / 2
        except ValueError:
            return float('nan')

df['total_sqft'] = df['total_sqft'].apply(convert_to_float).astype(float)
```

```python
new=df['total_sqft'].duplicated()
```

```python
df['duplicate_status_total_sqft'] = 'Not Duplicate'
df.loc[new, 'duplicate_status_total_sqft'] = 'Duplicate'
```

```python
features= df['total_sqft']
```

```python
def categorize_prices(prices, user_input):
    sorted_prices = sorted(prices)
    price_range = (sorted_prices[-1] - sorted_prices[0]) / 4

    categories = {
        'Category 1 (min-range)': f"{sorted_prices[0]}-{sorted_prices[0] + price_range}",
        'Category 2 (range)': f"{sorted_prices[0] + price_range + 1}-{sorted_prices[0] + 2 * price_range}",
        'Category 3 (2*range)': f"{sorted_prices[0] + 2 * price_range + 1}-{sorted_prices[0] + 3 * price_range}",
        'Category 4 (3*range-max)': f"{sorted_prices[0] + 3 * price_range + 1}-{sorted_prices[-1]}"
    }
    
    input_category = None
    user_price = float(user_input)

    if user_price <= sorted_prices[0] + price_range:
        input_category = 'Category 1 (min-range)'
    elif sorted_prices[0] + price_range < user_price <= sorted_prices[0] + 2 * price_range:
        input_category = 'Category 2 (range)'
    elif sorted_prices[0] + 2 * price_range < user_price <= sorted_prices[0] + 3 * price_range:
        input_category = 'Category 3 (2*range)'
    else:
        input_category = 'Category 4 (3*range-max)'

    return categories, input_category

user_input_price = input("Enter the Squerefoot: ")
result, user_category = categorize_prices(features, user_input_price)

for category, price_range in result.items():
    print(f"{category}: {price_range}")

print(f"User input belongs to: {user_category}")
```
