```python
## Example of the chat-gpt provide the some basic example to do how work this model and concept of the implementations
import pandas as pd
import spacy 
import csv
import spacy
import pandas as pd
import numpy as np
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
df = pd.read_csv("/home/codetrade/Downloads/CSV/row.csv")
df.head()
```

```python
entities = []
```

```python
for text in df['text']:
    doc = nlp(text)
    for ent in doc.ents:
        entities.append((ent.text, ent.label_))
```

```python
import warnings
warnings.filterwarnings("ignore")


entities
```

```python
df_entities = pd.DataFrame(entities, columns=['Text', 'Label'])
```

```python
df_entities.to_csv('output_of_Ner.csv',index=False)
```
