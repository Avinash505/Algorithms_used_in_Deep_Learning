```python
import spacy
import pandas as pd

nlp = spacy.load("en_core_web_sm")

# Assuming 'df' is a DataFrame with a column named 'text'
data = {"Text": [], "Label": []}

for text in df['text']:
    doc = nlp(text)
    for ent in doc.ents:
        data["Text"].append(ent.text)
        data["Label"].append(ent.label_)

# Create a DataFrame from the data
df_entities = pd.DataFrame(data)

# Count duplicate values based on the entity text and create a new column
df_entities['Duplicate_Text_Count'] = df_entities.groupby('Text')['Text'].transform('count')

# Count duplicate values based on the entity label and create a new column
df_entities['Duplicate_Label_Count'] = df_entities.groupby('Label')['Label'].transform('count')

# Print the DataFrame with duplicate counts
print(df_entities)

# Write the DataFrame to a CSV file
df_entities.to_csv("entity_counts.csv", index=False)

print("Entity counts written to: entity_counts.csv")
```
