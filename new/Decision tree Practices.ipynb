```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
X=df.iloc[:,:-1]
y=iris.target
```

```python
X,y
```

```python
##train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## By Default create the post pruning.
treemodel=DecisionTreeClassifier()
treemodel2=DecisionTreeClassifier(max_depth=2)
```

```python
treemodel.fit(X_train,y_train)
treemodel2.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel2,filled=True)
```

```python
y_pred=treemodel.predict(X_test)
y_pred2=treemodel2.predict(X_test)
```

```python
y_pred,y_pred2
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
score2=accuracy_score(y_pred2,y_test)
print(score)
print(score2)
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(classification_report(y_pred2,y_test))
```
